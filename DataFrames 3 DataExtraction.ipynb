# Second Arguments to loc and iloc Accessors

```python
bond = pd.read_csv("jamesbond.csv", index_col = "Film")
bond.sort_index(inplace = True)
bond.head()
```

```python
bond.loc["Moonraker", "Actor"] ## to extract data from the Row and the coloumn in order to locate the exact data to extract
bond.loc["Moonraker", "Director"]
bond.loc["Moonraker", "Budget"]
```

```python
bond.loc["Moonraker", ["Budget", "Actor"]] # If I want datas from two columns but from the same row
```

```python
bond.loc[["A View to a Kill", "Moonraker"], ["Budget", "Actor"]] # to extract to columns from two rows
```

```python
bond.loc["Moonraker", "Director":"Budget"]
```

```python
bond.loc["Moonraker":"Thunderball", "Director":"Budget"]
```

```python
bond.loc[:"Moonraker", "Director":"Budget"]
```

```python
bond.iloc[0]
bond.iloc[0:3, 3:5]
bond.iloc[[14, 25], 2:5]
bond.iloc[[14, 25], [2,4]]
bond.iloc[:7, [2,4]]
bond.iloc[:7, :3]
```
