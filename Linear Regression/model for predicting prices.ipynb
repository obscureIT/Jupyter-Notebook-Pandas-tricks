# Predicting prices of areas with linear Regression

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("homeprices.csv")
```

```python
df.head()
```

```python
%matplotlib inline #for drawing the plots
plt.xlabel('area')
plt.ylabel('price(US$+)')
plt.scatter(df.area, df.price, color="red", marker = "o")
```

```python
plt.scatter(df.area, df.price)
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[["area"]], df.price)
```

```python
reg.predict([[3300]])
```

# The linear regression model calculated the Coefficinet/slope/Gradient = m
# and the Intercept = b


```python
# Now lets check the coefficient
reg.coef_
```

```python
reg.intercept_
```

```python
# Since y = mx + b
# we have 
# m = 135.78767123
# b = 180616.43835616432
# Therefore for predicting x = 3300 as the area

135.78767123 * 3300 + 180616.43835616432 # Which should be the same as reg.predict([[3300]])
```

```python
# for predicting 5000
reg.predict([[5000]])
```

```python
# or we can ===>
135.78767123 * 5000 + 180616.43835616432
```

# Now to use this model to predict prices of areas from a different CSV File

```python
df2 = pd.read_csv("areas.csv")
```

```python
df2["predicted price"] = reg.predict(df2)
```
